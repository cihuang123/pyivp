### XYPoint
create a XYPoint object v1

```python
v1 = pyivp.XYPoint(30, 30)
```

### XYPattenBlock
create a XYPatternBlock object py_block, and add ID point v1

```python
py_block = pyivp.XYPatternBlock()
py_block.addIDPoint(v1)
```

you can set 5 parameters :  
1. block width 
2. block length 
3. swath width
4. angle 
5. id point

```python
py_block.setParam("block_width", 40)
```

```python
points = py_block.getLanePoints()
for point in points:
    print(point.get_spec(""))
```

```python
seglist = py_block.getLaneSegments()
for seg in seglist:
    print(seg.get_spec(""))
```

### 2 distance functions

```python
print(py_block.distanceToClosestEntry.__doc__)
```

```python
print(py_block.distanceToCrossAxis.__doc__)
```

```python
dist_entry = py_block.distanceToClosestEntry(osx = 10, osy = 30, any_segment = True) 
dist_crossAxis = py_block.distanceToCrossAxis(30, 35)
print(f'distance to entry : {dist_entry} \ndistance to cross axis : {dist_crossAxis}')
```
